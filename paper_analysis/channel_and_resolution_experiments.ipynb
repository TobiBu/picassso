# Channel experiments

```python
# histogram of global stellar properties
plt.figure(figsize=(20,10))
#gs = gridspec.GridSpec(1, 2, width_ratios=[1,1], height_ratios=[.6])
#gs.update(hspace=0.0, wspace=0.0)

bands = [5,4,2,1]
sigma_sm = np.asarray([0.09,0.163,0.111,0.152])
sigma_gm = np.asarray([0.128,0.150,0.132,0.154])
sigma_sz = np.asarray([0.143,0.219,0.183,0.22])
sigma_gz = np.asarray([0.195,0.227,0.187,0.212])
sigma_sa = np.asarray([0.086,0.165,0.115,0.125])
sigma_sfr = np.asarray([0.22,0.271,0.247,0.245])

mu_sm = np.asarray([0.103,0.031,0.069,0.081])
mu_gm = np.asarray([0.069,0.046,0.058,0.05])
mu_sz = np.asarray([0.158,0.127,0.135,0.158])
mu_gz = np.asarray([0.1,0.097,0.101,0.108])
mu_sa = np.asarray([0.028,0.089,0.063,0.063])
mu_sfr = np.asarray([0.082,0.07,0.121,0.114])

ax = plt.subplot(111)

ax.plot(bands,sigma_sm/sigma_sm[0],lw=5,label='stellar mass')
ax.plot(bands,sigma_sz/sigma_sz[0],lw=5,label='stellar metallicity')
ax.plot(bands,sigma_sa/sigma_sa[0],lw=5,label='stellar age')
ax.plot(bands,sigma_gm/sigma_gm[0],lw=5,ls='dashed', label='gas mass')
ax.plot(bands,sigma_gz/sigma_gz[0],lw=5,ls='dashed', label='gas metallicity')
ax.plot(bands,sigma_sfr/sigma_sfr[0],lw=5,ls='dashed', label='SFR')

ax.set_ylabel(r'$\sigma/\sigma_{ ugriz}$')
ax.set_xlabel(r'wavelength bands')
ax.set_xticks([1,2,4,5])
ax.set_xticklabels(['r','gr','ugri','ugriz'])
ax.legend(loc=0,ncol=2,fontsize=25)
#ax.set_xlim(9.4,12.25)
ax.set_ylim(0.925,2.05)


plt.savefig('./plots/band_comparison_sigma.pdf', bbox_inches='tight')
```

```python
# histogram of global stellar properties
plt.figure(figsize=(20,10))
#gs = gridspec.GridSpec(1, 2, width_ratios=[1,1], height_ratios=[.6])
#gs.update(hspace=0.0, wspace=0.0)

ax = plt.subplot(111)

ax.plot(bands,mu_sm/mu_sm[0],lw=5,label='stellar mass')
ax.plot(bands,mu_sz/mu_sz[0],lw=5,label='stellar metallicity')
ax.plot(bands,mu_sa/mu_sa[0],lw=5,label='stellar age')
ax.plot(bands,mu_gm/mu_gm[0],lw=5,ls='dashed', label='gas mass')
ax.plot(bands,mu_gz/mu_gz[0],lw=5,ls='dashed', label='gas metallicity')
ax.plot(bands,mu_sfr/mu_sfr[0],lw=5,ls='dashed', label='SFR')

ax.set_ylabel(r'$\mu/\mu_{ ugriz}$')
ax.set_xlabel(r'wavelength bands')
ax.set_xticks([1,2,4,5])
ax.set_xticklabels(['r','gr','ugri','ugriz'])
ax.legend(loc=2,ncol=2,fontsize=25)
#ax.set_xlim(9.4,12.25)
ax.set_ylim(0.2,3.75)


plt.savefig('./plots/band_comparison_mu.pdf', bbox_inches='tight')
```
